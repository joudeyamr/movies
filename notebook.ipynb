Data cleaning
<br>The process of fixing or removing incorrect, corrupted, incorrectly formatted, duplicate, or incomplete data within a dataset.


```python
#dropping columns i find unnessesary
df.drop(columns=['id', 'imdb_id', 'cast', 'homepage', 'tagline', 'keywords', 'overview','runtime'], inplace=True)
```

```python
#searching for duplicates
df[df.duplicated()]
```

```python
#found duplicates ,so decided to remove them as they dont add anything to the dataset
df.drop_duplicates(inplace=True)
```

```python
#making sure there is no value in budget is in negative
df = df[df['budget'] >= 0]
```

```python
#drop all nan values
df.dropna(inplace=True)
```

```python
#finding the unique values
df.nunique()
```

```python
#formatting henre to be easier to be read
formatted_genres = df.explode('genres')
formatted_genres.genres.head()
```

```python
#formatting production comoanies to be easier to read
formatted_prod_comp = df.explode('production_companies')
formatted_prod_comp.production_companies.head()
```

```python
#veiwing the first few row of the modified 'genre' column
df['genres'] = df['genres'].str.split('|')
df.genres.head()
```

```python
#viewing the first few rows of the modified 'production_companie' row
df['production_companies'] = df['production_companies'].str.split('|')
df.production_companies.head()
```

```python
#replacing zeros in the budget
df['budget'] = df['budget'].replace(0, df.budget.median())
```

```python
#making sure there isnt anymore zeros in the budget
zero_budget_count = df[df['budget'] == 0].shape[0]
print(zero_budget_count)
```

```python
#making a profit column
df["profit"] = df["revenue"] - df["budget"]
df["profit_adj"] = df["revenue_adj"] - df["budget_adj"]
```

```python
# Explode the 'genres' column to handle multiple genres
df_exploded = df.explode('genres')
```

```python
#formatting dates
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['release_year'] = df['release_year'].astype(int)
```

```python
#bar plot to visualise the answers
def groupby_plots(df: pd.DataFrame,
                  by: str,
                  column: str,
                  operation: str,
                  count:[int] = None) -> pd.Series:
                  operation_val = "Average" if operation == "mean" else "Total"
                  grouped = df.groupby(by)[column].agg(operation).sort_values(ascending=False)[:count]
                  plt.figure(figsize=(10, 6))
                  sns.barplot(x=grouped.index,y=grouped.values,palette="viridis")
                  plt.title(f'{operation_val} {column.title()} by {by.title()}')
                  plt.xlabel(f'{by.title()}')
                  plt.ylabel(f'{operation_val} {column.title()}')
                  plt.xticks(rotation=45, ha='right')
                  plt.tight_layout()
                  plt.show()
                  return grouped
```

```python
#writing a function to alwyas do a scatter plot with every answer
def groupby_scatterplot(df: pd.DataFrame,
                        by: str,
                        column: str,
                        operation: str,
                        count: int = None) -> pd.Series:
    operation_val = "Average" if operation == "mean" else "Total"
    grouped = df.groupby(by)[column].agg(operation).sort_values(ascending=False)[:count]
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=grouped.index, y=grouped.values, palette="viridis")
    plt.title(f'{operation_val} {column.title()} by {by.title()}')
    plt.xlabel(f'{by.title()}')
    plt.ylabel(f'{operation_val} {column.title()}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    return grouped
```

```python
#function to do correlations and then plot it
def plot_and_correlate(df: pd.DataFrame, x_col: str, y_col: str) -> float:
    correlation = df[x_col].corr(df[y_col])
    print(f"Pearson correlation coefficient between {x_col} and {y_col}: {correlation:.2f}")
    fig, ax = plt.subplots(figsize=(8, 6))
    sns.regplot(x=df[x_col], y=df[y_col], line_kws={"color": "r"}, ax=ax)
    ax.set_xlabel(x_col.capitalize())
    ax.set_ylabel(y_col.capitalize())
    ax.set_title(f"Relationship between {x_col.capitalize()} and {y_col.capitalize()}")
    plt.show()
    return correlation
```

<h3>Limitations:</h3>

<p>2.Handling duplicates:</p>
<p>Although duplicate entries were eliminated, it's likely that some were overlooked or that some valid submissions were mistakenly classified as duplicates.</p>
<p>3.Handling genre and production companies:</p>
<p>In order to manage many entries, the genres and production_companies columns were exploded. However, this approach considers each genre and production company equally, ignoring the principal genre or main production firm, which could distort the study.</p>
<p>4.Analyising correlations:</p>
<p>A linear relationship between variables is assumed by the Pearson correlation coefficient. Accurately capturing non-linear correlations between profit and budget may be problematic.</p>
<p>5.Level of accuracy in time data:</p>
